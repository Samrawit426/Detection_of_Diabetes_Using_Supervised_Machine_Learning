```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/diabetes.csv')

print(df.head())
print(df.info())
# Summary statistics for numerical features
summary_stats = df.describe()

print(summary_stats)
# Save summary statistics to CSV
summary_stats.to_csv('../results/summary_statistics.csv')





```

```python
# Class distribution
sns.countplot(x='Outcome', data=df)
plt.title('Class Distribution')

# Save plot
plt.savefig("../results/class_distribution.png", dpi=300)
plt.show()
```

```python
# Feature distributions
features = ['Glucose', 'BMI', 'Age']
df[features].hist(bins=20, figsize=(10, 6))
plt.suptitle('Feature Distributions')

# Save plot
plt.savefig("../results/feature_distributions.png", dpi=300)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')

# Save plot
plt.savefig("../results/correlation_heatmap.png", dpi=300)
plt.show()
```
